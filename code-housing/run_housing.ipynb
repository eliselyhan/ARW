```python
# Data:
# https://www.dubaipulse.gov.ae/data/dld-transactions/dld_transactions-open

# load data
df_raw = pd.read_csv(path + '/Transactions.csv')
encoding = 'label'
df = preprocess(df = df_raw, encoding = encoding)


prop = [0.6, 0.2]
window_sizes_train = [1, 4, 16, 64, 256]
window_sizes_val = [1, 4, 16, 64, 256]
list_t = np.arange(1, 192)

list_seed = np.arange(1, 21) * 1000

for seed in list_seed:
    print('seed {} started'.format(seed))

    data = splitting(df = df, prop = prop, seed = seed)
    MSE_ARW, MSE_fixed = test_online(data, list_t, window_sizes_train, window_sizes_val, seed)

    np.savetxt(path + '/results_housing/MSE_ARW_{}_{}_{}_full.txt'.format(seed, prop[0], prop[1]), MSE_ARW)
    np.savetxt(path + '/results_housing/MSE_fixed_{}_{}_{}_full.txt'.format(seed, prop[0], prop[1]), MSE_fixed)

    print('seed {} finished'.format(seed))

```
